```python
import pandas as pd
import random
```

```python
pacientes = pd.read_excel("paciente.xlsx")
```

```python
pacientes = pacientes[pacientes['provincia']=='GUAYAS']
```

```python
parroquias = pd.read_excel("parroquias_ecuador.xls")
```

```python
def get_parroquia(indice):
    parroquias  = ['Ayacucho','Bolívar','Carbo','Febres Cordero','García Moreno','Letamendi' ,'Nueve de Octubre','Olmedo','Roca','Rocafuerte','Sucre','Tarqui','Urdaneta','Ximena']    
    return parroquias[indice]
```

```python
pacientes["parroquia"] = pacientes["parroquia"].apply(lambda x: get_parroquia(random.randint(1,13)).upper() if pd.isnull(x) else x.upper())
```

```python
pacientes["direccion"] = 1
```

```python
pacientes[["parroquia","direccion"]].groupby(["parroquia"]).count()
```
